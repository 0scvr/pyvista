```python
import vtki
from vtki import examples
```

```python
dataset = examples.load_uniform()
# dataset.set_active_scalar('Spatial Point Data')
dataset
```

```python
dataset.plot()
```

```python
thresh = dataset.threshold([100, 500])
thresh.plot(notebook=True, show_bounds=True)
```

```python
slc = dataset.slice()
slc.plot()
```

```python
clp = dataset.clip(invert=False)
clp.plot()
```

```python
iso = dataset.contour()
iso.plot()
```
